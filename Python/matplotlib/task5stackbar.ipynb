```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year

year_category_sales = df.groupby(["Year", "Category"])["Sales"].sum().unstack()
years = year_category_sales.index.astype(str)
categories = year_category_sales.columns
x = np.arange(len(years))
bottom = np.zeros(len(years))
for category in categories:
   plt.bar(x, year_category_sales[category], bottom=bottom, label=category)
   bottom += year_category_sales[category].values
plt.title("Stacked Sales by category per year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(x, years)
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```
