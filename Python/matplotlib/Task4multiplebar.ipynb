```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")
df
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year
year_category_sales = df.groupby(["Year", "Category"])["Sales"].sum().unstack()
categories = year_category_sales.columns
years = year_category_sales.index
x = np.arange(len(categories))
width = 0.2
```

```python
for i, year in enumerate(years):
    plt.bar(x + i * width, year_category_sales.loc[year], width=width, label=str(year))
plt.title("Sales by Category for Different years")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(x + width, categories)
plt.legend()
plt.tight_layout()
plt.show()
```
