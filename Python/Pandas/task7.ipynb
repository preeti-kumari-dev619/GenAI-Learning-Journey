```python
import pandas as pd
```

###python


```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}
df = pd.DataFrame(students)
```

```python
print("Average Marks Per Subjects")
print(df.groupby('Subject')['Marks'].mean())
```

```python
print("Count of Students per Subject:")
print(df.groupby('Subject')['Name'].count())
```

```python
print("Maximum Marks per Subject:")
print(df.groupby('Subject')['Marks'].max())
```
