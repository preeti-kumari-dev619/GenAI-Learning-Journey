```python
import pandas as pd
```

```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}
df = pd.DataFrame(students)
```

```python
print("students who score more than 75")
print(df[df['Marks'] > 75])
```

```python
print("Students belonging to Math:")
print(df[df['Subject'] == 'Math'])
```

```python
print("Students who scored more than average marks")
avg = df['Marks'].mean()
print(f"Avg Marks: {avg}")
print(df[df['Marks'] > avg])
```

```python
print("Students who Failed (Marks < 70):")
print(df[df['Marks'] < 70])
```
