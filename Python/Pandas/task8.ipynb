```python
import pandas as pd
```

```python
students = {
    'Name': ['Amit', 'Rahul', 'Neha', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}
df = pd.DataFrame(students)
```

```python
df.plot(kind='bar', x='Name', y='Marks', title='Student Marks - Bar Graph', legend='False', color='steelblue')
```

```python
s = df['Marks']
```

```python
s.plot(kind='line', title='Marks - Line Graph')
```

```python
s.plot(kind='hist', title='Marks - Histogram')
```
