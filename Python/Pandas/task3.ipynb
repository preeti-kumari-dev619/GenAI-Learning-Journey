```python
import pandas as pd
```

```python
marks = pd.Series([78, 85, 90, 66, 72])
```

```python
print(f"Maximum Marks: {marks.max()}")
```

```python
print(f"Minimum Marks : {marks.min()}")
```

```python
print(f"Sum of Marks: {marks.sum()}")
```

```python
print(f"Mean marks: {marks.mean()}")
```

```python
print("Pass/Fail (>=70)")
passed = marks.apply(lambda x: "Pass" if x >= 70 else "Fail")
print(passed)
```

```python
print("\nNumber of students passed:", (marks >= 70).sum())
```
