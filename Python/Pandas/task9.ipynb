```python
import pandas as pd
```

```python
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}
```

```python
df = pd.DataFrame(sales)
```

```python
print(f"Total Revenue: {df['Revenue'].sum()}")
```

```python
print(f"Average Daily Revenue: {df['Revenue'].mean()}")
```

```python
print(df.loc[df['Revenue'].idxmax(), 'Day'])
```

```python
avg = df['Revenue'].mean()
print("Days Revenue where average")
print(df[df['Revenue'] > avg])
```

```python
df.plot(kind='bar', x='Day', y='Revenue', title='Revenue vs Day', legend='False')
```
