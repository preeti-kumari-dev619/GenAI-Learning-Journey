```python
import pandas as pd
```

```python
students = {
    'Name': ['Amint', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subjects': ['Math', 'Math', 'Science', 'Science', 'Math']
}
df = pd.DataFrame(students)
```

```python
print("df.info()")
df.info()
```

```python
print("df.describe()")
df.describe()
```

```python
print("df.head()")
print(df.head())
```

```python
print("df.tail()")
print(df.tail())
```

```python
print("sorted by marks(decending)")
df_sorted = df.sort_values(by='Marks', ascending=False)
print(df_sorted)
```

```python
print("After resetting index:")
df_sorted_reset = df_sorted.reset_index(drop=True)
print(df_sorted_reset)
```
