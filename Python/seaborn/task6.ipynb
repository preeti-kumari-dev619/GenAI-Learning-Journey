```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df = df.dropna(subset=["Age"])
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="Pclass",
    y="Fare"
)
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="Sex",
    y="Age"
)
plt.title("Age Distribution By Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(
    data=df,
    x="Pclass",
    y="Fare"
)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x="Sex"
)
plt.title("Number of Male and Female Passenger")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x="Sex",
    hue="Survived"
)
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
