```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df = df.dropna(subset=["Age", "Fare"])
```

```python
plt.figure(figsize=(8,6))
sns.histplot(
    data=df,
    x="Age",
    y="Fare"
)
plt.title("Bivariate Histogram: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    data=df,
    x="Age",
    y="Fare",
    fill=True
)
plt.title("Bivariate KDE plot: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    data=df,
    x="Age",
    y="Fare"
)
plt.title("Bivariate KDE Contour Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
