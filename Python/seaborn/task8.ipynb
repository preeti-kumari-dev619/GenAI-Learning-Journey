```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df = df.dropna(subset=["Age", "Fare"])
```

```python
g = sns.FacetGrid(
    data=df,
    col="Sex",
    height=5,
    aspect=1.2
)
g.map_dataframe(
    sns.scatterplot,
    x="Age",
    y="Fare"
)
g.fig.suptitle("FacetGrid: Age vs Fare by Sex", y=1.05)
plt.show()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    col="Sex",
    kind="scatter",
    height=5,
    aspect=1
)
plt.show()
```

```python
sns.catplot(
    data=df,
    x="Pclass",
    y="Fare",
    hue="Sex",
    col="Survived",
    kind="bar",
    height=5,
    aspect=1
)
plt.show()
```

```python
sns.displot(
    data=df,
    x="Age",
    col="Sex",
    kde=True,
    height=5,
    aspect=1
)
plt.show()
```
