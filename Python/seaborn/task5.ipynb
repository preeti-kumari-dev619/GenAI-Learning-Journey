```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.dropna(subset=["Age", "Fare"])
```

```python
sns.pairplot(
    data=df,
    vars=["Age", "Fare", "Pclass", "Survived"],
    hue="Sex"
)
plt.show()
```

```python
corelation = df[["Age", "Fare", "Pclass", "Survived"]].corr()
print(corelation)
```

```python
plt.figure(figsize=(8,6))
correlation = df[["Age", "Fare", "Pclass", "Survived"]].corr()
sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```
