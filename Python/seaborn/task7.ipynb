```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df = df.dropna(subset=["Age", "Fare"])
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    data=df,
    x="Age",
    y="Fare"
)
plt.title("Regression Plot: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    data=df,
    x="Age",
    y="Fare",
    marker="o"
)
plt.title("Regression Plot with Circle Markers")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    data=df,
    x="Age",
    y="Fare",
    ci=None
)
plt.title("Regression Plot Without Confidence Interval")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
