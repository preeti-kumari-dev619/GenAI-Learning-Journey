```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Sex"
)
plt.show()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Sex",
    kind="scatter"
)
plt.show()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Sex",
    kind="scatter",
    height=6,
    aspect=1.5
)
plt.show()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Sex",
    kind="scatter",
    height=6,
    aspect=1.5
)
plt.title("Relationship between Age and Fare")
plt.show()
```
