```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df = df.dropna(subset=["Age"])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df,
    x="Age",
    y="Fare"
)
plt.title("Line Plot between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df,
    x="Age",
    y="Fare",
    marker="o"
)
plt.title("Scatter Style line plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.relplot(
    data=df,
    x="Age",
    y="Fare",
    kind="line",
    col="Sex",
    height=5,
    aspect=1.2
)
plt.show()
```
