```python
import numpy as np
```

```python
sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])
```

```python
total_sales = np.sum(sales)
```

```python
print(f"Total Weekly Sales: {total_sales}")
```

```python
avg_sales = np.mean(sales)
```

```python
print(f"Average Daily Sales: {avg_sales}")
```

```python
highest_day = np.argmax(sales) + 1
```

```python
lowest_day = np.argmin(sales) + 1
```

```python
print(f"Highest Sales : Day , {highest_day} = {np.max(sales)}")
```

```python
print(f"Lowest Sales : Day , {lowest_day} = {np.min(sales)}")
```

```python
std_sales = np.std(sales)
```

```python
print(f"Standard Deviation: {std_sales}")
```

```python
above_avg_sales = np.where(sales > avg_sales)[0] + 1
```

```python
print(f"Days Above Average: {above_avg_sales}")
```
