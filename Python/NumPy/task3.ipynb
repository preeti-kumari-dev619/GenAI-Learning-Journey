```python
import numpy as np
```

```python
values = np.array([2, 4, 6, 8, 10])
```

```python
print(f"Square Root: {np.sqrt(values)}")
```

```python
print(f"Exponential: {np.exp(values)}")
```

```python
print(f"Nature log: {np.log(values)}")
```

```python
print("Cumulative sum:", np.cumsum(values))
```

```python
print("Sum of all elements:", np.sum(values))
```
