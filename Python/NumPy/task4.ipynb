```python
import numpy as np
```

```python
data = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
```

```python
print(f"Row wise Sum: {np.sum(data, axis=1)}")
```

```python
print(f"Column wise sum: {np.sum(data, axis=0)}")
```

```python
print(f"Minimum Value: {np.min(data)}")
```

```python
print(f"Maximum Value: {np.max(data)}")
```

```python
print(f"Mean Value: {np.mean(data)}")
```
