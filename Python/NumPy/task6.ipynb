```python
import numpy as np
```

```python
marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])
```

```python
sorted_marks = np.sort(marks)
```

```python
print(f"Sorted Array: {sorted_marks}")
```

```python
p25 = np.percentile(marks, 25)
```

```python
p50 = np.percentile(marks, 50)
```

```python
p75 = np.percentile(marks, 75)
```

```python
print(f"25th percentile: {p25}")
```

```python
print(f"50th percentile: {p50}") 
```

```python
print(f"75th percentile: {p75}")
```

```python
average = np.mean(marks)
```

```python
above_avg_count = np.sum(marks > average)
```

```python
print(f"Average Marks: {average}")
```

```python
print(f"above_avg_count: {above_avg_count}")
```
