```python
import numpy as np
```

```python
marks = np.array([78, 85, 90, 62, 72, 88, 95, 60])
```

```python
print(f"Mean: {np.mean(marks)}")
```

```python
print(f"Median: {np.median(marks)}")
```

```python
print(f"Variance: {np.var(marks)}")
```

```python
print(f"Standard Deviation: {np.std(marks)}")
```

```python
print(f"Min: {np.min(marks)}")
```

```python
print(f"Max: {np.max(marks)}")
```

```python
print(f"Range: {np.max(marks) - np.min(marks)}")
```
